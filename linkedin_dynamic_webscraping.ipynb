```python
#initiate all lists
offer_url = []
title_list = []
company_name_list = []
location_list = []
workplace_list = []
applicants_list = []
posting_date_list = []
employment_type_list = []
senority_level_list = []
number_of_employees_list = []
sector_list = []
application_through_linkedIn_list = []
python_required_list = []
promoted_list = []
number_of_followers_list = []
state_list = []
```

```python
#initializing a count for the pages
count = 2
#setting a loop for the pages
while count <= 41:
    
    time.sleep(2)
    source = BeautifulSoup(driver.page_source, "html.parser")
    
    #getting tags of titles which have offer url within them
    all_elements = source.find_all("a", class_ = "disabled ember-view job-card-container__link job-card-list__title")
    
    #to scroll six times setting a range to get to bottom of the page
    scroll_times = range(6)
    
    #loop to scroll to load the entire page
    for x in scroll_times:
        time.sleep(1)
        #reti
        store_id = all_elements[-1]["id"]
        locate_element = driver.find_element(By.ID, store_id)
        driver.execute_script("arguments[0].scrollIntoView(true);", locate_element)
        time.sleep(2)
        source = BeautifulSoup(driver.page_source, "html.parser")
        all_elements = source.find_all("a", class_ = "disabled ember-view job-card-container__link job-card-list__title")
    
    if len(all_elements) != 25:
        for x in scroll_times:
            time.sleep(1)
            #reti
            store_id = all_elements[-1]["id"]
            locate_element = driver.find_element(By.ID, store_id)
            driver.execute_script("arguments[0].scrollIntoView(true);", locate_element)
            time.sleep(1)
            source = BeautifulSoup(driver.page_source, "html.parser")
            all_elements = source.find_all("a", class_ = "disabled ember-view job-card-container__link job-card-list__title")
        
    
    footer_elements = source.find_all("ul", class_ = "job-card-list__footer-wrapper job-card-container__footer-wrapper flex-shrink-zero display-flex t-sans t-12 t-black--light t-normal t-roman")
    print(len(footer_elements))
    for x in footer_elements:
        if x.find("li", class_ = "t-12 t-normal t-black--light job-card-container__footer-item") == None:
            promoted = False
    
        else:
            promoted = True
        promoted_list.append(promoted)
        
    
    all_elements = source.find_all("a", class_ = "disabled ember-view job-card-container__link job-card-list__title")
    print(len(all_elements))
    for x in all_elements:
        new_url = "https://www.linkedin.com" +  x["href"]
        offer_url.append(new_url)
        
        detail_listing = driver.find_element(By.ID, x["id"])
        detail_listing.click()
        
        time.sleep(2)
        source_new = BeautifulSoup(driver.page_source, "html.parser")
    
        #title
        time.sleep(1)
        title = source_new.find("h2", class_ = "t-24 t-bold jobs-unified-top-card__job-title").text.strip()
        title_list.append(title)
    
        #company name
        if source_new.find("a", class_ = "ember-view t-black t-normal") == None:
            company_name = None
        else:
            company_name = source_new.find("a", class_ = "ember-view t-black t-normal").text.strip()
        company_name_list.append(company_name)
    
        #location
        location = source_new.find("span", class_ = "jobs-unified-top-card__bullet").text.strip()
        location_list.append(location)
    
        #workplace
        if source_new.find("span", class_ = "jobs-unified-top-card__workplace-type") == None:
            workplace = "Other"
    
        else:
            workplace = source_new.find("span", class_ = "jobs-unified-top-card__workplace-type").text.strip()
        workplace_list.append(workplace)
    
        #posting date
        try:
            posting_date = source_new.find("span", class_ = "jobs-unified-top-card__posted-date").text.strip()
            posting_date_list.append(posting_date)
        except:
            posting_date = None
            posting_date_list.append(posting_date)
    
        #applicants
        if source_new.find("span", class_ = "jobs-unified-top-card__applicant-count") == None:
            applicants = None
        else:
            applicants = int(source_new.find("span", class_ = "jobs-unified-top-card__applicant-count").text.strip().split(" ")[-2])
        applicants_list.append(int(applicants))
    
    
    
        all_lists = source_new.find_all("li", class_ = "jobs-unified-top-card__job-insight")
        count1 = 0
        count2 = 0
        count3 = 0
        for x in all_lists:
            if x.find("li-icon") == source_new.find("li-icon", {"type" : "job"}):
                store = x.text.split("·")
            
                count1 = 1
                employment_type = store[0].strip()
        
                if len(store) ==  1:
                    senority_level = None
                else:
                    senority_level = store[1].strip()
                
                
            if x.find("li-icon") == source_new.find("li-icon", {"type" : "company"}):
                store = x.text.split("·")
                temp = store[0].strip().split(" ")[0]
                print(temp)
                temp2 = temp.split("-")[0]
                print(temp2)
                number_of_employees = "".join(re.findall(r'\d+', temp2))
                count2 = 1
        
                if len(store) ==  1:
                    sector = None
                else:
                    sector = store[1].strip()
            
            try:
                if x.find("li-icon") == source_new.find("li-icon", {"class" : "ivm-view-attr__icon--signal-positive"}):
                    store = x.find("li-icon")["aria-label"]
                    count3 = 1
                    if store == "Actively recruiting":
                        state = "On-going"
                    elif store == "Be an Early Applicant":
                        state = "Early Applications"
                    else:
                        state = "Others"
            except:
                if count3 == 0:
                    state = "Others"
        
        
        
        if count1 == 0:
            employment_type = None
            senority_level = None
        
        if count2 == 0:
            number_of_employees = None
            sector = None
        

            
        employment_type_list.append(employment_type)
        senority_level_list.append(senority_level)
        number_of_employees_list.append(int(number_of_employees))
        sector_list.append(sector)
        state_list.append(state)
        
        
        if source_new.find("button", class_ = "jobs-apply-button artdeco-button artdeco-button--3 artdeco-button--primary ember-view") == None:
            application_through_linkedIn = False
        else:
            application_through_linkedIn = True
        application_through_linkedIn_list.append(application_through_linkedIn)
        
        
        
        description = source_new.find("div", class_ = "jobs-description__content jobs-description-content").text.strip().lower()
        
        if "python" in description:
            python_required = True
        else:
            python_required = False
            
        python_required_list.append(python_required)
        
        
        time.sleep(1)
        detail_window_tag = driver.find_element(By.XPATH, '//*[@id="SALARY"]')
        driver.execute_script("arguments[0].scrollIntoView(true);", detail_window_tag)
        
        
        time.sleep(1)
        fresh_source = BeautifulSoup(driver.page_source, "html.parser")
        if fresh_source.find("div", class_ = "artdeco-entity-lockup__subtitle ember-view t-16") == None:
            number_of_followers = None
        else:
            followers = fresh_source.find("div", class_ = "artdeco-entity-lockup__subtitle ember-view t-16").text.strip().split(" ")[0]
            number_of_followers = int("".join(re.findall(r'\d+', followers)))
        
        number_of_followers_list.append(int(number_of_followers))
        
        time.sleep(1)
        
    
    
    
    try:
        id_of_button = source_new.find("li" , {"data-test-pagination-page-btn" : str(count)})["id"]
        x_path_build = "//*[@id='" + str(id_of_button) + "']/button"
        next_page_button = driver.find_element(By.XPATH, x_path_build)
        next_page_button.click()
        print("Page Complete")
        
    except:
        try:
            id_of_button = source_new.find("li" , {"data-test-pagination-page-btn" : str(count - 1)})["id"]
            number_of_id = int("".join(re.findall(r'\d+', id_of_button))) + 1
            id_button_build = "ember" + str(number_of_id)
            x_path_build = "//*[@id='" + str(id_button_build) + "']/button"
            next_page_button = driver.find_element(By.XPATH, x_path_build)
            next_page_button.click()
            print("Page Complete")
        except:
            print(1)
    
        
    
    count += 1
    
    
        
        
  
    
    
    
    
    
    
    
    
```
